```python
import os
import time
from PyPDF2 import PdfReader
import pymongo
from bson import ObjectId
from datetime import datetime

# MongoDB connection
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["kiosk"]
collection = db["costing"]

# Function to count pages in a PDF file
def count_pdf_pages(pdf_path):
    try:
        reader = PdfReader(pdf_path)
        return len(reader.pages)
    except Exception as e:
        print(f"Error counting pages in PDF: {e}")
        return 0

# Function to simulate user input for location and printing type selection
def get_user_input():
    # Fetch locations and printing types from MongoDB
    locations = [loc['name'] for loc in collection.find({"field_name": "Location"})]
    printing_types = collection.find({"field_name": "Printing type"})
    binding_options = collection.find({"field_name": "Binding and Finishing"})
    
    # Ask for Location (Folder Path for PDFs)
    folder_path = r"pdfs\\"

    # Get all PDF files from the folder
    pdf_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.pdf')]
    total_pages = 0
    num_pdfs = len(pdf_files)
    
    # Count total pages in all PDFs
    for pdf_file in pdf_files:
        pdf_path = os.path.join(folder_path, pdf_file)
        pages = count_pdf_pages(pdf_path)
        total_pages += pages
        print(f"PDF: {pdf_file}, Pages: {pages}")
    
    # Display location options
    print("\nSelect a Location:")
    for idx, location in enumerate(locations, 1):
        print(f"{idx}. {location}")
    location_choice = int(input("Enter the number for the location: ")) - 1
    selected_location = locations[location_choice]
    print(f"Selected Location: {selected_location}")

    # Display printing types options
    print("\nSelect a Printing Type:")
    printing_type_dict = {}
    for idx, print_type in enumerate(printing_types, 1):
        print_type_name = print_type['name']
        print_type_cost = print_type['cost']
        printing_type_dict[idx] = {'name': print_type_name, 'cost': print_type_cost}
        print(f"{idx}. {print_type_name} - Cost: {print_type_cost}")
    printing_choice = int(input("Enter the number for Printing Type: "))
    selected_printing_type = printing_type_dict[printing_choice]
    print(f"Selected Printing Type: {selected_printing_type['name']} - Cost per page: {selected_printing_type['cost']}")

    # Calculate printing cost
    total_printing_cost = total_pages * selected_printing_type['cost']

    # Display binding options
    print("\nSelect a Binding and Finishing Type:")
    binding_dict = {}
    for idx, binding in enumerate(binding_options, 1):
        binding_dict[idx] = {'name': binding['name'], 'cost': binding['cost']}
        print(f"{idx}. {binding['name']} - Cost: {binding['cost']}")
    binding_choice = int(input("Enter the number for Binding and Finishing: "))
    selected_binding = binding_dict[binding_choice]
    print(f"Selected Binding and Finishing: {selected_binding['name']} - Cost per PDF: {selected_binding['cost']}")

    # Calculate binding cost
    total_binding_cost = selected_binding['cost'] * num_pdfs

    # Ask for user details
    name = input("\nEnter your Name: ")
    phone = input("Enter your Phone: ")
    email = input("Enter your Email: ")
    description = input("Enter a Description: ")
    transaction_id = input("Enter Transaction ID: ")

    # Calculate total cost
    total_cost = total_printing_cost + total_binding_cost

    # Print the summary
    timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    print("\nReceipt:")
    print(f"Time Stamp: {timestamp}")
    print(f"Name: {name}")
    print(f"Phone: {phone}")
    print(f"Email: {email}")
    print(f"Description: {description}")
    print(f"Transaction ID: {transaction_id}")
    print(f"Total PDFs: {num_pdfs}")
    print(f"Total Pages: {total_pages}")
    print(f"Selected Printing Type: {selected_printing_type['name']}")
    print(f"Cost Per Page of Selected Printing Type: {selected_printing_type['cost']}")
    print(f"Selected Location: {selected_location}")
    print(f"Selected Location Cost: 0 (Locations have no direct cost)")
    print(f"Selected Binding and Finishing: {selected_binding['name']}")
    print(f"Selected Binding and Finishing Cost (per PDF): {selected_binding['cost']}")
    print(f"Total Price: {total_cost}")

if __name__ == "__main__":
    get_user_input()
```
